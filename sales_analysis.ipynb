```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. ফাইল লোড করা এবং ডেটা ক্লিনিং ---

# আপনার ফোল্ডারের পরম পাথ ব্যবহার করা হয়েছে
file_path = "G:/sales-dashboard-project/data/sales.csv" 

try:
    # নিশ্চিত সমাধানের জন্য explicit delimiter (sep=',') ব্যবহার করা হলো
    df = pd.read_csv(file_path, sep=',')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print(f"Error: sales.csv not found at {file_path}")
    exit()

# কলামের নামগুলি পরিষ্কার করা এবং বাধ্যতামূলকভাবে সেট করা
df.columns = df.columns.str.strip() 
df.columns = ['Date', 'Product', 'Category', 'Region', 'Units Sold', 'Unit Price']
print("Column names successfully standardized!")

# --- 2. ফিচার ইঞ্জিনিয়ারিং (Revenue কলাম তৈরি) ---

# তারিখের কলামকে সঠিক ডেটাটাইপে (datetime) রূপান্তর
df['Date'] = pd.to_datetime(df['Date']) 
# Revenue কলাম তৈরি
df['Revenue'] = df['Units Sold'] * df['Unit Price']

print("\n--- Key Performance Indicators (KPIs) ---")

# --- 3. KPI গণনা এবং আউটপুট ---
total_revenue = df['Revenue'].sum()
total_units = df['Units Sold'].sum()
top_product = df.groupby('Product')['Revenue'].sum().idxmax()
top_region = df.groupby('Region')['Revenue'].sum().idxmax()

print(f"Total Revenue: {total_revenue}")
print(f"Total Units Sold: {total_units}")
print(f"Top Product: {top_product}")
print(f"Top Region: {top_region}")
print("------------------------------------------")

# --- 4. ট্রেন্ড বিশ্লেষণ ---
df['Month'] = df['Date'].dt.month
monthly_revenue_trend = df.groupby('Month')['Revenue'].sum()

# লাইন চার্ট প্লট করা
plt.figure(figsize=(10, 5))
monthly_revenue_trend.plot(kind='line', marker='o')
plt.title("Monthly Revenue Trend (Month 1 = Jan, Month 2 = Feb)")
plt.xlabel("Month")
plt.ylabel("Revenue (in USD)")
plt.grid(True)
plt.show()

# --- 5. ডেটা সেভ করা (গুরুত্বপূর্ণ ধাপ) ---
# এখন যেহেতু Power BI বন্ধ, ফাইলটি সফলভাবে সেভ হবে।

output_path = "G:/sales-dashboard-project/data/sales.csv" 
df.to_csv(output_path, index=False)
print(f"\n✅ Updated dataset with Revenue column successfully saved to: {output_path}")
```
